```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(0,2 * math.pi, 0.1)
f1 = [5*math.sin(i) for i in x]
f2 = [2*math.sin(2*i) for i in x]
f3 = [1*math.sin(3*i) for i in x]
plt.plot(x, f1)
plt.plot(x, f2, 'ro')
plt.plot(x, f3, 'g--')
plt.show()
```
