```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Set size and gray background color
fig = plt.figure()
fig.set_size_inches(10,4)
axes = fig.add_subplot(1, 1, 1, axisbg='#333333')

# Force vertical zoom out (hackish!)
axes.plot([14,14],[2.01, -2.01], 'go')

x = np.linspace(-.1, 4*math.pi+.1, 10000)
y = 1 * np.sin(x)
y2 = 1 * np.sin(2*x)
y3 = 2 * np.sin(x)
_ = axes.plot(x,y, 'orange', x, y2, 'cyan', x, y3, 'red')
```
